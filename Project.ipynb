## **part b**

> build the model

```r
model <- lm(col7 ~ col1 + col2 + col3 + col4 + col5 + col6, my_data[sample, ])
```

> summarising the model

First, for all **coefficients** (including the intercept), we have the pointwise estimation and **standard devivation** which helps to find interval estimation for the coefficient. Biside those, for a specific significant level, the **t-value** is given to check whether the feature is even useful in regression at all.

**Residuals** are the distance of the points from the line and we also see a summary of them, including minimum, maximum, median, 1Q and 3Q, for a good regression line we expect the resudials to be close to zero, so these values (especially median, 1Q and 3Q) should be close to zero, however maximum and minimum are highly affected by outliers.

And finally, the three parameters, **Residual standard error, Multiple R-squared, F-statistic** are metrics how good our data fits the the regression line, for example, the greater the **F-statistics** is, the better the data fit the model. **R-squared** is always between 0 and 1, zero indicates that the model explains none of the variability of the response data around its mean, 1 indicates that the model explains all the variability of the response data around its mean.


```r
summary(model)
```

> size of coefficient and comparing them

As we see excluding the intercept, coefficient related to **column two** has the highest absolute value, and coefficient related to **column five** has the lowest abosolute value, **however this comparison is incorrect**, since different features have different sacles and the coefficient value is directly related to the scale and same changes of value in different features - for instance weight and blood pressure - is completely a different thing.

**The alternative solution** is using t-values, which can explain how much the target function is dependent on the feature, hence the best feature above is **column three** (having highest t-value), and the worst feature in the model is **column six** (having the least t-value) typically any t-value outside of interval $[-2, 2]$ is acceptable.

## **part c**

> threshold of zero

```r
# prediction for test set
pred <- predict.lm(model, test_feature)
# assign negative numbers to -1 and non-negative ones to +1
pred[pred >= 0] = 1
pred[pred < 0] = -1
```

> compute accuracy

```r
(length(pred) - sum(abs(pred - test_label)) / 2) / length(pred)
```

> plot accuracy for different thresholds


```r
# make an empty list for accuracies
accuracies = c()
# compute accuracy for different thresholds
for (e in seq(-2, 2, 0.01)){
    # prediction for test set
    pred <- predict.lm(model, test_feature)
    # assign negative numbers to -1 and non-negative ones to +1
    pred[pred >= e] = 1
    pred[pred < e] = -1
    accuracies = c(accuracies, (length(pred) - sum(abs(pred - test_label)) / 2) / length(pred))
}
```

```r
accuracies
```

```r
plot(seq(-2, 2, 0.01), accuracies, type = "l", xlab = 'threshold', col = "blue", lty = 2, lwd = 2.5)
```

> regression using one-hot encoding

```r
one_hot <- function(index) {
    region <- rep(0, 22)
    region[index] <- 1
    return (region)
}
```

```r
# make an initial dataframe   
r <- rep(0, 22)
year <- price <- c(0)
df <- data.frame(r[1], r[2], r[3], r[4], r[5], r[6], r[7], r[8], r[9], r[10], r[11], r[12], r[13], 
                 r[14], r[15], r[16], r[17], r[18], r[19], r[20], r[21], r[22], year, price)
# iteration
for (row in rownames(houses)) {
    for (col in colnames(houses)) {
        df[nrow(df) + 1,] = c(one_hot(as.numeric(row)), as.numeric(col), as.numeric(houses[row, col]))
    }
}
# remove the first extra row
df <- df[-1,]
```

```r
model = lm(price ~ r.1. + r.2. + r.3. + r.4. + r.5. + r.6. + r.7. + r.8. + r.10. + r.11. + r.12. + 
           r.13. + r.14. + r.15. + r.16. + r.17. + r.18. + r.19. + r.20. + r.21. + r.22. + year, df)
```

```r
summary(model)
```

> coclusion

Generally speaking, one-hot encoding is better than using the region number directly since the region numbers order should not be important in our model, because even using a relabling, the result shouldn't be changed much, also pay attention to residual amounts:


$$\begin{cases}\mbox{First Model}&:\mbox{Residual standard error: 50290 on 899 degrees of freedom}\\\\\mbox{Second Model}&:\mbox{Residual standard error: 47640 on 879 degrees of freedom}\end{cases}$$

the slight decrease of residual error also show that the second model is performing better too, comparing R-squared errors, it can be seen that:


$$\begin{cases}\mbox{First Model}&:\mbox{Adjusted R-squared:  0.4699}\\\\\mbox{Second Model}&:\mbox{Adjusted R-squared:  0.5243}\end{cases}$$

we typically know when $0.3<r<0.5$ the model isn't performing well an when $0.5 <r<0.7$ the model show moderate effect size 

Blindly comparing F-statistics the second model isn't doint better but since we're taking more features (almost 10 times more) then it's fine!

```r
t = (mean(xi) - mean(yi)) / sqrt((length(xi) - 1) * (var(xi) + var(yi)) * 2 / length(xi))
p = pt(q = t, df = 2 * length(xi) - 2, lower.tail = T)
```

```r
sprintf("p_value is equal to %f", p)
```

Comparing the p-value to the our significant level (for example $5$%), we see we cannot reject the null hypothesis

```r
sprintf('the mean: %f and the variance: %f', mean(zi), var(zi))
```
